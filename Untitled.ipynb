```python
import json
import geojson
import pandas as pd
```

```python
df_mapper = pd.read_csv("data.csv")
mapper = dict(zip(df_mapper.id, df_mapper.name))
```

```python

with open('train_labels/xView_train.geojson', 'r') as f:
    gj = geojson.load(f)
```

```python
images_info = {}

for poly in gj['features']:
    coord = poly["geometry"]["coordinates"]
    image = poly["properties"]["image_id"]
    poly_type = mapper.get(poly["properties"]["type_id"], None)
    if not poly_type:
        continue
    info = {"polygon_coordinates": coord, "polygon_type":poly_type}
    
    if image not in images_info:
        images_info[image] = [info]
    else:
        images_info[image].append(info)
```

```python
for key, content in images_info.items():
    with open(f"annotations/{key}.json", "w") as f:
        json.dump(content, f, indent=4)
```
